```python
#  Setup Mediapipe Hands
mpHands = mp.solutions.hands
hands = mpHands.Hands(
    static_image_mode=False,
    max_num_hands=1,
    min_detection_confidence=0.7,
    min_tracking_confidence=0.5
)
mpDraw = mp.solutions.drawing_utils
```

```python
def fingers_up(lmList):
    """
    Checks which fingers are extended.
    Returns a list of 5 integers (1 for up, 0 for down).
    Order: [Thumb, Index, Middle, Ring, Pinky]
    """
    tips = [4, 8, 12, 16, 20]
    fingers = []

    # Thumb Logic (Horizontal check)
    if lmList[tips[0]][0] > lmList[tips[0]-1][0]:
        fingers.append(1)
    else:
        fingers.append(0)

    # Other 4 Fingers Logic (Vertical check)
    for i in range(1, 5):
        if lmList[tips[i]][1] < lmList[tips[i]-2][1]:
            fingers.append(1)
        else:
            fingers.append(0)

    return fingers



print("Sucess")
```

<h3 style="color:#38bdf8; text-align:center;">Finger Detection Logic</h3>

<p style="font-size:14px;">
This function detects which fingers are <b>UP</b> or <b>DOWN</b> using hand landmark positions.
</p>

<pre style="background:#020617; padding:10px; border-radius:6px; color:#38bdf8;">

</pre>

<ul style="font-size:14px; line-height:1.6;">
<li> Returns list → [Thumb, Index, Middle, Ring, Pinky]</li>
<li> Thumb uses horizontal (X-axis) comparison</li>
<li> Other fingers use vertical (Y-axis) comparison</li>
<li> 1 = Finger UP, 0 = Finger DOWN</li>
</ul>

```python
# Initial State
cap = cv2.VideoCapture(0)
mode = "NONE"
last_switch = 0
control_active = False
last_toggle = 0

print("Launching Camera... Press 'ESC' to close.")

while True:
    success, frame = cap.read()
    if not success: break

    # Flip the frame for a 'mirror' effect
    frame = cv2.flip(frame, 1)
    h, w, c = frame.shape
    
    # Create a black background for the UI (Optional: Use 'frame' instead of 'img' to see yourself)
    img = np.zeros_like(frame) 

    # Processing hand landmarks
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(rgb)

    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            lmList = []
            for id, lm in enumerate(handLms.landmark):
                cx, cy = int(lm.x * w), int(lm.y * h)
                lmList.append((cx, cy))

            fingers = fingers_up(lmList)
            current_time = time.time()

            # CONTROL TOGGLE
            # Fist [0,0,0,0,0] = ON | Open Palm [1,1,1,1,1] = OFF
            if fingers == [0,0,0,0,0] and current_time - last_toggle > 1:
                control_active = True
                last_toggle = current_time
            elif fingers == [1,1,1,1,1] and current_time - last_toggle > 1:
                control_active = False
                last_toggle = current_time

            # MODE SWITCH
            # Index + Middle = VOLUME
            if fingers == [0,1,1,0,0] and current_time - last_switch > 1:
                mode = "VOLUME"
                last_switch = current_time
            # Index + Middle + Ring = BRIGHTNESS
            elif fingers == [0,1,1,1,0] and current_time - last_switch > 1:
                mode = "BRIGHTNESS"
                last_switch = current_time

            # CALCULATION
            # Distance between Thumb (4) and Index (8)
            x1, y1 = lmList[4]
            x2, y2 = lmList[8]
            length = math.hypot(x2 - x1, y2 - y1)

            # UI Visuals
            cv2.circle(img, (x1, y1), 10, (255,0,0), -1)
            cv2.circle(img, (x2, y2), 10, (255,0,0), -1)
            cv2.line(img, (x1,y1), (x2,y2), (0,255,0), 3)

            # EXECUTION 
            if control_active:
                if mode == "VOLUME":
                    vol = np.interp(length, [30, 200], [minVol, maxVol])
                    volume.SetMasterVolumeLevel(vol, None)
                    per = np.interp(length, [30, 200], [0, 100])
                    cv2.putText(img, "VOLUME MODE", (20,50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)
                
                elif mode == "BRIGHTNESS":
                    bright = np.interp(length, [30, 200], [0, 100])
                    sbc.set_brightness(int(bright))
                    per = bright
                    cv2.putText(img, "BRIGHTNESS MODE", (20,50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,0), 2)
                
                bar = np.interp(length, [30, 200], [400, 150])
            else:
                bar, per = 400, 0

            # UI Progress Bar
            status_color = (0,255,0) if control_active else (0,0,255)
            status_text = "CONTROL: ON" if control_active else "CONTROL: OFF"
            cv2.putText(img, status_text, (350,50), cv2.FONT_HERSHEY_SIMPLEX, 1, status_color, 2)
            cv2.rectangle(img, (50,150), (85,400), (255,255,255), 2)
            cv2.rectangle(img, (50, int(bar)), (85,400), (0,255,0), -1)
            cv2.putText(img, f'{int(per)} %', (40,450), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)

            mpDraw.draw_landmarks(img, handLms, mpHands.HAND_CONNECTIONS)

    cv2.imshow("AI Gesture UI", img)
    if cv2.waitKey(1) & 0xFF == 27: break

cap.release()
cv2.destroyAllWindows()


print("Complete")
```
