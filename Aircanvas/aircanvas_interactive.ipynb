```python

# ---------- CONFIG ----------
WIDTH, HEIGHT = 640, 480 #To change system configrations 
CONFIDENCE = 0.7

tool = "BRUSH"
colors = [(255,0,255),(0,255,0),(255,255,0),(0,0,255)]
color_ptr = 0
brush_thickness = 8

# ---------- MEDIAPIPE ----------
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(
    max_num_hands=1,
    model_complexity=0,
    min_detection_confidence=CONFIDENCE,
    min_tracking_confidence=CONFIDENCE
)

# ---------- VARIABLES ----------
canvas = np.zeros((HEIGHT, WIDTH, 3), dtype=np.uint8)
xp, yp = 0, 0
cloc_x, cloc_y = 0, 0
prev_time = 0

# ---------- CAMERA ----------
cap = cv2.VideoCapture(0)
cap.set(3, WIDTH)
cap.set(4, HEIGHT)

# ---------- HAND CONNECTIONS ----------
connections = [
    (0,1),(1,2),(2,3),(3,4),
    (0,5),(5,6),(6,7),(7,8),
    (0,9),(9,10),(10,11),(11,12),
    (0,13),(13,14),(14,15),(15,16),
    (0,17),(17,18),(18,19),(19,20)
]

# ---------- FUNCTIONS ----------
def get_fingers(lm):
    fingers = []
    fingers.append(lm[4].x < lm[3].x)
    for tip in [8,12,16,20]:
        fingers.append(lm[tip].y < lm[tip-2].y)
    return fingers

def draw_skeleton(img, lm):
    points = []
    for p in lm:
        x = int(p.x * WIDTH)
        y = int(p.y * HEIGHT)
        points.append((x,y))

    for c in connections:
        cv2.line(img, points[c[0]], points[c[1]], (0,255,255), 2)

    for i in [0,4,8,12,16,20]:
        cv2.circle(img, points[i], 5, (0,255,0), -1)

def draw_ui(bg):
    cv2.rectangle(bg, (0,0), (WIDTH,60), (30,30,30), -1)
    cv2.putText(bg, f"{tool}", (10,40),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)

    for i, col in enumerate(colors):
        thickness = 6 if i == color_ptr else 2
        cv2.rectangle(bg, (150+i*70,10), (210+i*70,50), col, -1)
        cv2.rectangle(bg, (150+i*70,10), (210+i*70,50), (255,255,255), thickness)

def handle_buttons(x,y):
    global color_ptr
    if y < 60:
        for i in range(4):
            if 150+i*70 < x < 210+i*70:
                color_ptr = i

# ---------- MAIN LOOP ----------
while True:
    success, frame = cap.read()
    if not success:
        break

    frame = cv2.flip(frame,1)

    # Reduce size for faster processing
    small = cv2.resize(frame, (0,0), fx=0.5, fy=0.5)
    rgb = cv2.cvtColor(small, cv2.COLOR_BGR2RGB)
    result = hands.process(rgb)

    bg = np.zeros((HEIGHT, WIDTH, 3), dtype=np.uint8)
    draw_ui(bg)

    if result.multi_hand_landmarks:
        for hand_lms in result.multi_hand_landmarks:
            lm = hand_lms.landmark

            draw_skeleton(bg, lm)

            x = int(lm[8].x * WIDTH)
            y = int(lm[8].y * HEIGHT)

            cloc_x += (x - cloc_x) / 2
            cloc_y += (y - cloc_y) / 2

            fingers = get_fingers(lm)

            # All fingers → clear canvas
            if all(fingers):
                canvas[:] = 0

            # Thumb only → eraser
            elif fingers[0] and not any(fingers[1:]):
                tool = "ERASER"

            # Two fingers → select color
            elif fingers[1] and fingers[2]:
                xp, yp = 0, 0
                handle_buttons(int(cloc_x), int(cloc_y))
                tool = "BRUSH"

            # One finger → draw
            elif fingers[1] and not any(fingers[2:]):
                if xp == 0 and yp == 0:
                    xp, yp = cloc_x, cloc_y

                color = (0,0,0) if tool=="ERASER" else colors[color_ptr]
                thickness = 25 if tool=="ERASER" else brush_thickness

                cv2.line(canvas,
                         (int(xp), int(yp)),
                         (int(cloc_x), int(cloc_y)),
                         color, thickness)

                xp, yp = cloc_x, cloc_y
            else:
                xp, yp = 0, 0

    # FPS Calculation
    curr = time.time()
    fps = 1/(curr-prev_time) if prev_time!=0 else 0
    prev_time = curr

    cv2.putText(bg,f'FPS:{int(fps)}',(10,HEIGHT-10),
                cv2.FONT_HERSHEY_SIMPLEX,0.7,(0,255,0),2)

    # Merge drawing + UI
    output = cv2.addWeighted(bg,1,canvas,1,0)

    # 🔥 MINI CAMERA VIEW
    small_cam = cv2.resize(frame, (200,150))
    output[HEIGHT-150:HEIGHT, WIDTH-200:WIDTH] = small_cam

    cv2.rectangle(output,
                  (WIDTH-200, HEIGHT-150),
                  (WIDTH, HEIGHT),
                  (255,255,255), 2)

    cv2.imshow("Air Canvas + Skeleton", output)

    key = cv2.waitKey(1)
    if key == 27 or key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
